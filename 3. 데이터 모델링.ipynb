### 데이터 불러오기

```python
import numpy as np

train_q1_data= np.load(open('quora-question-pairs/q1_train.npy','rb'))
train_q2_data= np.load(open('quora-question-pairs/q2_train.npy','rb'))
train_labels= np.load(open('quora-question-pairs/label_train.npy','rb'))
```

```python
train_input= np.stack((train_q1_data, train_q2_data),axis=1)
```

```python
train_input.shape
```

```python
from sklearn.model_selection import train_test_split

## 검증 데이터셋 생성
train_input, eval_input, train_label, eval_label= train_test_split(train_input,train_labels, test_size=0.2, random_state=731)
```

```python
import xgboost as xgb

train_data= xgb.DMatrix(train_input.sum(axis=1), label=train_label)
eval_data= xgb.DMatrix(eval_input.sum(axis=1), label=eval_label)

data_list= [(train_data, 'train'), (eval_data, 'valid')]
```

```python
data_list
```

```python
params= {}
params['objective']= 'binary:logistic'
params['eval_metric']= 'rmse'

bst= xgb.train(params, train_data, num_boost_round= 1000, evals=data_list, early_stopping_rounds=10)
```

```python
import numpy as np

test_q1_data= np.load(open('quora-question-pairs/q1_test.npy','rb'))
test_q2_data= np.load(open('quora-question-pairs/q2_test.npy','rb'))
test_id_data= np.load(open('quora-question-pairs/test_id.npy','rb'), allow_pickle=True)
```

```python
test_input= np.stack((test_q1_data, test_q2_data), axis=1)
test_data= xgb.DMatrix(test_input.sum(axis=1))
test_predict= bst.predict(test_data)
```

```python
import pandas as pd

output= pd.DataFrame({'test_id': test_id_data, 'is_duplicate': test_predict})
```
